```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
dataset = pd.read_csv("input.csv")
```

```python
dataset
```

```python
riasec = ["Realistic", "Investigate", "Artistic", "Social", "Enterprising", "Conventional"]
learning_style = ["Linguistic", "Logical-mathematical", "Spatial", "Bodily-Kinesthetic", "Musical", "Interpersonal", "Intrapersonal", "Naturalistic",]
```

```python
dataset[riasec] = dataset[riasec]/7 *100
dataset[learning_style] = dataset[learning_style]/8 *100
```

```python
X = dataset.iloc[:, 0:18].values
y = dataset.iloc[:, 18].values
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
#from sklearn.preprocessing import StandardScaler

#sc = StandardScaler()
#X_train = sc.fit_transform(X_train)
#X_test = sc.transform(X_test)

#X_train
```

```python
from sklearn.ensemble import RandomForestClassifier

regressor = RandomForestClassifier(n_estimators=100, random_state=0)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
filename = 'random_forest_model.sav'
pickle.dump(regressor, open(filename, 'wb'))
```

```python
fn=np.array(dataset.columns[:-1])
cn=["STEM", "HUMSS", "ABM", "GAS"]

fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (10,10), dpi=1020)
tree.plot_tree(regressor.estimators_[0],
               feature_names = fn, 
               class_names=cn,
               filled = True);
fig.savefig('rf_individualtree.png')
```
