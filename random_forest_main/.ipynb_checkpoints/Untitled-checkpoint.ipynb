```python
dataset = pd.read_csv("input.csv")

print(dataset.shape)

dataset.head()
```

```python
# Investigate all the elements whithin each Feature 

for column in dataset:
    unique_vals = np.unique(dataset[column])
    nr_values = len(unique_vals)
    if nr_values < 36:
        print('The number of values for feature {} :{} -- {}'.format(column, nr_values,unique_vals))
    else:
        print('The number of values for feature {} :{}'.format(column, nr_values))
```

```python
# Visualize the data using seaborn Pairplots
g = sns.pairplot(dataset, hue = 'CLASS', diag_kws={'bw': 0.2})
```

```python
# Your code goes here
X = dataset.drop('CLASS', axis=1).values# Input features (attributes)
y = dataset['CLASS'].values # Target vector
print('X shape: {}'.format(np.shape(X)))
print('y shape: {}'.format(np.shape(y)))

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, test_size=0.3, random_state=0)
```

```python
dt = DecisionTreeClassifier(criterion='entropy', max_depth=2, random_state=1)
dt.fit(X_train, y_train)
```

```python
import graphviz 

dot_data = tree.export_graphviz(dt, out_file=None, 
    feature_names=dataset.drop('CLASS', axis=1).columns,    
    class_names=dataset['CLASS'].unique().astype(str),  
    filled=True, rounded=True,  
    special_characters=True)
graph = graphviz.Source(dot_data)
graph
```
